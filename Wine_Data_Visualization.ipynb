## Part-1:
## Question-1: 
Which pair of variables are highly correlated (leave out the class label - first column)?

## Answer-1:
Total phenols and Flavanoids are highly correlated. The values between Total phenols and Flavanoids (0.86 and 0.86) are close to 1.

```python
import pandas as pd
import numpy as np
import seaborn as sb
import matplotlib.pyplot as plt
%matplotlib inline

wine_data_df = pd.read_csv("https://archive.ics.uci.edu/ml/machine-learning-databases/wine/wine.data", sep=',', 
                  names = ['Class label', 'Alcohol', 'Malic acid', 'Ash', 'Alcalinity of ash ', 'Magnesium', 
                           'Total phenols', 'Flavanoids', 'Nonflavanoid phenols', 'Proanthocyanins', 
                           'Color intensity', 'Hue', 'OD280/OD315 of diluted wines', 'Proline'])

wine_data_df.head()
wine_data_df.drop('Class label', axis=1, inplace=True)
wine_data_df_corr = wine_data_df.corr()

fig, ax = plt.subplots(figsize=(15, 10))
sb.heatmap(round(wine_data_df_corr, 2).iloc[:, :], annot=True, linewidths=.5, ax=ax, cmap="rainbow",fmt='.2f')
```

## Question-2: 
Are there any pair of variables that have a non-linear relationship?

## Answer-2:
Yes. "Ash" and "OD280/OD315 of diluted wines" have a non-linear relationship. If we see the above plot we can that the values between "Ash" and "OD280/OD315 of diluted wines" are 0 or their correlation is 0. And the plot below, shows that the regression line is straight and parallel to x-axis which also means no correlation.

```python
sb.set(style="ticks")

sb.lmplot(x='Ash', y='OD280/OD315 of diluted wines',data=wine_data_df, fit_reg=True)
```

## Question-3: 
When comparing wine cultivar type 1,2, and 3, Do any of the type(s) have datapoints with high hue and low color intensity?

## Answer-3:
Yes. Class ’type 2’ has datapoints with high hue and low color intensity.

```python
from pandas.plotting import parallel_coordinates

df = pd.read_csv("https://archive.ics.uci.edu/ml/machine-learning-databases/wine/wine.data", sep=',', 
                  names = ['Class label','Alcohol', 'Malic acid','Ash','Alcalinity of ash ','Magnesium',
                           'Total phenols','Flavanoids','Nonflavanoid phenols','Proanthocyanins',
                           'Color intensity','Hue','OD280/OD315 of diluted wines','Proline '])

parallel_coordinates(df, class_column='Class label', cols=['Hue', 'Color intensity'])
parallel_coordinates
```
