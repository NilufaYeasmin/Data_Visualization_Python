## Updated Chart

```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.DataFrame({'Label0':[1, 5, 5, 5, 12, 23, 49], 'Name':['Sophisticated Visual Aids (1%)', 'Opening Joke (5%)', 'Spontaneous Moment (5%)', 
                                                             'Statement of Utter Certainty (5%)', 'Snappy Refrain (12%)', 'Personal Failure (23%)', 'Contrarian Thesis (49%)']}, 
                  columns=['Label0', 'Name'])
labels = ['Sophisticated Visual Aids (1%)', 'Opening Joke (5%)', 'Spontaneous Moment (5%)', 
          'Statement of Utter Certainty (5%)', 'Snappy Refrain (12%)', 'Personal Failure (23%)', 'Contrarian Thesis (49%)']
colors = ['yellow', 'blue', 'green', 'lightskyblue', 'orange', 'red', 'pink']

fig, ax = plt.subplots(figsize=(10, 10))
fig.subplots_adjust(0.3, 0, 1, 1)

ax.pie(df['Label0'], labels=df['Name'], colors=colors, startangle=45)
ax.axis('equal')
plt.title("\n Success Factor of TED Talk", fontsize=20)
plt.legend(loc='upper left', labels=labels, prop={'size': 12}, bbox_to_anchor=(0.0, 1), bbox_transform=fig.transFigure)
plt.show()
```

## Updated Image

```python
from pygantt.pygantt import gantt
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.patches as mpatches
%matplotlib inline

df = pd.DataFrame({'Start':[221, 172, 203, 184, 183, 167, 180, 143, 165, 157], 
                   'Finish': [239, 231, 218, 198, 199, 197, 195, 192, 179, 172],
                   'Name':['Lead Software Engineer Contractor', 'Directors', 'Product Management Director', 'Human Resource Director',
                           'Engineering Director', 'Staff User Experience Designer', 'Senior Partner Technology Manager',
                           'Senior Managers', 'Marketing Director', 'Group Product Manager'],
                   'Type': ['A', 'A', 'A', 'B', 'B', 'B', 'B', 'B', 'C', 'C']},                  
                  columns=['Start', 'Finish', 'Type', 'Name'])


fig = plt.figure(figsize=(20, 10))

gantt(task=df.Name, start=df.Start, finish=df.Finish, task_type=df.Type, color={"A":"steelblue", "B":"lightblue","C":"firebrick"})

plt.title("\n Top 10 salaries at Google based on Job Category" ,fontsize= 18) 
plt.grid(b=bool, which='both', axis='both')
plt.xlabel('Annual Salary in thousands of Dollars', fontsize=18)
plt.ylabel('Job Category', fontsize=18)

plt.show()
```

## Summary of changes:
* If we see the image, we can find that there are starting and ending based on job category. From the updated horizontal bar chart or Gantt chart we can visualize the insight data more clearly . 

### Chart title:
* The Chart title has been changed to Top 10 slaries at Google based on Job Category.
