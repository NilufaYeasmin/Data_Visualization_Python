# Histogram

## Age of students in a remote school in India were listed as follows: 3, 11, 12, 19, 22, 23, 24, 25, 27, 29, 15, 6, 27, 17, 29.

```python
import pandas as pd
import numpy as np
import matplotlib
import matplotlib.pyplot as plt
matplotlib.style.use('ggplot')
```

```python
age = [3,11,12,19,22,23,24,25,27,29,15,6,27,17,29]
```

```python
plt.hist(age, bins=20,color='r',density= False,edgecolor= 'k',histtype='bar',alpha= 0.5)
plt.title('My Own Histogram')
plt.xlabel('Age of the students')
plt.ylabel('Frequency')
plt.show()
```

# **Why visualization is appropriate for the data: 

### I used Histogram to represent distribution of continuous data. Histogram plots quantitative data with ranges of the data into group.This data set represents a distribution of continuous data(numerical data) and I can summarize full data sets graphically by using Histogram. So, I think Histogram is the appropriate visualization for the data.  

# Bar graph

## 150 students of class were asked to state their favorite subjects and these were the results of survey.

```python
import matplotlib.pyplot as plt; plt.rcdefaults()
import numpy as np
import matplotlib.pyplot as plt
 
objects = ('French', 'English', 'Maths', 'Geography', 'Science')
y_sub = np.arange(len(objects))
result = [30,20,26,38,34]
 
plt.bar(y_sub, result, width=0.5, align='center', color='b',edgecolor= 'k', alpha=0.5)
plt.xticks(y_sub, objects)
plt.ylabel('Result of Survey')
plt.title('150 students Favorite subjects')
 
plt.show()
```

# ** Why visualization is appropriate for the data:

### I used Bar graph to  compare the result of several subjects since bar graph (each column) represents a group defined by a categorical variable. I've compared different subjects or to track the most favorite subject of 150 students by comparing different subject.I think Bar Plot is the appropriate visualization for the data.


# A company collected statistics about children to show that children liked TV more than other activities. 
These were the results of the survey

```python
import matplotlib.pyplot as plt

# Pie chart, where the slices will be ordered and plotted counter-clockwise:
labels = 'School', 'Sleep', 'Playing', 'Study', 'T.V.', 'Others'
sizes = [6,8,2,4,1,3]
explode = (0, 0, 0, 0,0,0)  

fig1, ax1 = plt.subplots()
ax1.pie(sizes, explode=explode, labels=labels, autopct='%1.1f%%',
        shadow=True, startangle=90)
ax1.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.

plt.show()

```

Q.4: A clinal trial measurement of two different groups A and B were reported. Which group is performing better (high values are better)?

```python
import pandas as pd
import numpy as np
import matplotlib
import matplotlib.pyplot as plt
matplotlib.style.use('ggplot')

A = {0.22, -0.87, -2.39, -1.79, 0.37, -1.54, 1.28, -0.31, -0.74, 1.72, 0.38, -0.17, -1.62, -1.10, 0.30, 0.15, 2.30, 0.19, -0.50, -0.09}
B = {-5.13, -2.15, -3.43, -3.83, 0.50, -3.25, 4.32, 0.63, 5.18, -0.43, 7.11, 4.87, -3.10, -5.81, 3.76, 6.31, 2.58, 0.07, 5.76, 3.50}

list_a=[]

for item in A:
    list_a.append(item)
#print(list_a)
#print(len(list_a))
list_b=[]
for item in B:
    list_b.append(item)
#print(len(list_b))
fig = plt.figure()
ax1 = fig.add_subplot(121)
ax1.title.set_text('A')
ax1.boxplot(list_a)


ax2 = fig.add_subplot(122)
ax2.title.set_text('B')
ax2.boxplot(list_b)

plt.show()
```

# **Why visualization is appropriate for the data

### I used Box Plot to represent the measurements of the two groups since boxplot shows the distribution of data in different quartiles. In our case we have same amount of data for group A and B; and high values are considered better. As we can see in the above boxplots, the median for B is above 0 that is higher than the median of A. It means 50% of the measurements of group B are higher than A, and hence 50% of the measurements of group B are better than A. Furthermore, the Q3 and Q4 measures in the boxplot for group B are significantly higher than the A. However, the Q1 measurements are better in boxplot for group A than B, i.e., the smallest 25% of the measurements of group A data is better (higher value) than the group B. Therefore, except Q1, the other quartiles in the boxplot for B are better than A given that higher values represent better performance. Considering all the above, we can conclude group B is performing better than group A.
